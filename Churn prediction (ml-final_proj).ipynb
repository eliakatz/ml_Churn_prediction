```python
# read the data file, Returns dataframe
# (imports: import pandas as pd)

def read_file(filename, index_col=None, parse_dates=None):
    # read file and create dataframe
    try:
        df = pd.read_csv(filename +'.csv',index_col=index_col, parse_dates=parse_dates)
    except:
        df = pd.read_csv('/datasets/' + filename +'.csv', index_col=index_col, parse_dates=parse_dates)   
    return df
#================================================================================

# read a portion of the data file, Returns dataframe
# (imports: import pandas as pd)

def read_portion(filename, num_of_lines):
    # read file and create dataframe
    try:
        df = pd.read_csv(filename +'.csv', nrows=num_of_lines)
    except:
        df = pd.read_csv('/datasets/' + filename +'.csv', nrows=num_of_lines)   
    return df
#================================================================================

# read an optimized data file, Returns dataframe
# (imports: import pandas as pd)
def read_optim_file(filename, col_type=None, col_dates=False, num_of_rows=None, index_col=None): #, parse_dates=None):
    try:
        df = pd.read_csv(filename +'.csv',
                         nrows=num_of_rows,
                         dtype=col_type,  # example: {'Device': 'category'},
                         parse_dates=col_dates,   # example: ['Start Ts', 'End Ts'],
#                         index_col=index_col, 
#                         parse_dates=parse_dates,
                         ) 
    except:
        df = pd.read_csv('/datasets/' + filename +'.csv',
                         nrows=num_of_rows,
                         dtype=col_type,  # example: {'Device': 'category'},
                         parse_dates=col_dates,   # example: ['Start Ts', 'End Ts'],
#                         index_col=index_col, 
#                         parse_dates=parse_dates,
                         ) 
    return df
#================================================================================

# show info, head, tail, sample.
def show_info(df):
    # show dataframe info
    display(df.info())
    
    # show dataframe head, tail, sample
    display(df.head(10))
    display(df.tail(10))
    display(df.sample(10))

#================================================================================
# read a portion of the data file, show info. Returns dataframe

def check_portion(filename, num_of_lines=500):
    df = read_portion(filename, num_of_lines)
    display(df.info(memory_usage='deep'))
    return df
#================================================================================

# prepear the data

def rename_columns(df):
    #rename columns
    df.columns=df.columns.str.lower()
    df.columns=df.columns.str.replace(' ','_')
    display(df)

#================================================================================
#check nulls

def missing_values(data):
    miss_values = data.isnull().sum().to_frame()
    miss_values = miss_values.rename(columns = {0: 'missing_values'})
    miss_values['%_of_total'] = ((miss_values['missing_values'] / data.shape[0]) *100).round(2)
    display(miss_values)

#================================================================================
# prepear the data

def prepare_data(df):
    print('Rename columns:')
    rename_columns(df)
    print('Missing values summary:')
    missing_values(df)
    print('Display value_counts:')
    check_value_count(df)
    
#================================================================================
# check the value_counts in the columns to find categorial columns

def check_value_count(df):
    for c in df.columns:
        display(df[c].value_counts())
```

## Abstract:
* People leave services for a variety of reasons. Some of them are evitable and some are not.

* In order to make an efficient use of the personnel's time and the company's budget, the company should focus on the evitable cases.

* Once a person decides to leave, it's harder and more expensive trying to keep him, than avoiding this situation from the beginning.

* Communication service is not something people like to change often (especially if it requires time loss, change of peripherals etc.), and usually the decision doesn't come out of the blue. So, once the customer leaves, he is not likely to be back (soon or at all), beside the bad impact it might have on other customers (family, friend etc.) as well.

The table describes some main reasons for churn:

| |Reasons for churn|Is there anything to avoid that|
|:---:|:--- | :---: |
|1.|Costs | yes |
|2.|Bad service| yes|
|3.|Competitors have a better offer|yes|
|4.|Temporary reasons|yes|
|5.|Lack of adequate infrastructure | no |
|6.|Customer's moving to place in which Interconnect's service is not available | no | 
|7.|Customer's moving to place in which Interconnect's service is available | yes | 
|8.|Customer passed away | no |
|9.|Customer no longer needs the service | no |

#### Change type of 'totalcharges'

```python
contract['totalcharges_num'] = pd.to_numeric(contract['totalcharges'], errors='coerce', downcast= 'float')
contract.head()
```

```python
contract.info()
```

```python
#change dtype to 'float64' and round to 2 decimal digits.
contract['totalcharges_num']=contract['totalcharges_num'].astype('float64').round(2)
contract.info()
```

```python
# find the non numeric records (didn't appear as nulls, and prevented the change of dtype)
contract[contract['totalcharges_num'].isna()]
```

###### Findings:
* The empty strings in 'totalcharges' belong to the new customers (begindate = 1/2/2020). These customers will be excluded from the training since they don't have any use for prediction or for test. #### Add column 'left' (This will be the target column (replacing 'enddate'))

```python
contract['left'] = (contract['enddate']!= 'No').astype('category')
contract.head()
```

#### Change type of 'enddate', and replace 'No' with the report's day (1/2/2020) for 'duration' calculations.

```python
# replace 'No' with the last date of the dataset (1/2/2020)
contract['enddate'] = contract['enddate'].replace('No', '2020-02-01')
contract.head()
```

```python
contract['enddate'] = pd.to_datetime(contract['enddate'])
contract.info()
```

#### Add columns 'beginmonth', 'beginyear'  and 'duration_d'

```python
contract['beginmonth'] = contract['begindate'].dt.month
contract['beginyear'] = (contract['begindate']).dt.year
contract['duration_d'] = (contract['enddate'] - contract['begindate'])
contract.head()
```

```python
contract['duration_d_int'] = contract['duration_d'].dt.days
contract['duration_d_int']
```

```python
# calculate the duration in months (the datetime class converts 30 days to 0 months).
contract['duration_m'] = (contract['duration_d_int']/30.0)
contract['duration_m'] = contract['duration_m'].round()
```

```python
# check the records of duration=0
contract[contract['duration_m']==0]
```

* Only the new customers have now duration=0.

## Model

```python
# print the auc_roc and accuracy
def find_model_aucroc(model, features_train, target_train, features_valid, target_valid):
    
    model.fit(features_train, target_train)
    predicted_valid = model.predict(features_valid)
    
    probabilities_valid = model.predict_proba(features_valid)
    probabilities_one_valid = probabilities_valid[:, 1]

    auc_roc = skm.roc_auc_score(target_valid, probabilities_one_valid)   
    accuracy_valid = accuracy_score(target_valid, predicted_valid) 

    print('auc_roc: ', auc_roc)
    print('accuracy: ', accuracy_valid)
```

```python
# upsample
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    
    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)

# shuffle 
    features_upsampled, target_upsampled = shuffle(features_upsampled, target_upsampled, random_state=STATE)
    
    return features_upsampled, target_upsampled 
```

```python
# downsample
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=STATE)] + [features_ones]) 
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=STATE)] + [target_ones]) 
    features_downsampled, target_downsampled = shuffle(features_downsampled, target_downsampled, random_state=STATE)
    
    return features_downsampled, target_downsampled
```

```python
def check_upsample(model, features_train, features_valid):
    for i in range(2,10):
        print('repeat =', i)
        features_upsampled, target_upsampled = upsample(features_train, target_train, i)

        find_model_aucroc(model, features_upsampled, target_upsampled, features_valid, target_valid)
        print()    
```

```python
def check_downsample(model, features_train, features_valid):
    for i in range(1, 8):
        print('repeat =', i/10)
        features_downsampled, target_downsampled = downsample(features_train, target_train, i/10)

        find_model_aucroc(model, features_downsampled, target_downsampled, features_valid, target_valid)
        print()
```

### What difficulties did you encounter and how did you manage to solve them?

* The 'begindate' and duration ('enddate'-'begindate') seems to be a good prediction (according to the graphs id EDA). But, they cannot be used together as features, because they cause target leakage (they result the 'enddate' - the target).


* In order to solve this problem, two sets of features were created. One is based on the 'begindate', dropping all the duration features. The other one is based on duration, dropping the begindate (and the derivatives: 'beginmonth' and 'beginyear') features.

### What were some of the key steps to solving the task?

* Plotting the graphs in the EDA step, to see the differences graphically. In general, it looks like the less committed to the company, the higher is the churn. But, it is a case of 'the chicken and the egg' - did the customers leave because they wanted a temporary service from the beginning (temporary relocation, special event etc.), or did they leave due to poor service or a better offer from a competitor.


* Deciding if it's a case of time series or not. Finding that the earliest 'enddate' is 4 months ago, made it clear that this is not the case. There is no data for the train set, no trends or seasonality to check.


* Creating two sets of features to avoid target leakage.


* Bypassing a bug when converting the duration in days to duration in months - the 'datetime' converted 30 days to 0 months, and spoiled the data.


* Splitting the data to groups: active, left and new for comparison.


* Dropping the new customers from the data.


* Comparing the total charge with the calculated one. It's something that in the real world worth checking. Except one, the only customers who were charged as expected (monthly * duration) were those who are just one month as customers. That means that either there's an accountability  problem, or a frequent change in tariffs or contracts, or offers to attract the customers.
